# Probando Polylla con vem

Primero se definen las funciones a probar

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
from mpl_toolkits.axes_grid1 import make_axes_locatable
from vem import vem, plot_solution

def square_domain_boundary_condition(points):
    x = points[:, 0]
    y = points[:, 1]

    g = (1 - x) * y * np.sin(np.pi * x)

    return g

def square_domain_rhs(point):
    x = point[0]
    y = point[1]

    f = 15 * np.sin(np.pi * x) * np.sin(np.pi * y)
    return f

def test_domain_rhs(point):
    return 0

def test_boundary_condition(points):
    x = points[:, 0]
    y = points[:, 1]

    g = x + y
    return g

def test2_domain_rhs(point):
    return 0

def test2_boundary_condition(points):
    x = points[:, 0]
    y = points[:, 1]

    g = np.exp(x)*np.sin(y)
    return g
    

# pa pruebas
def harmonic_boundary_condition(points):
    x = points[:, 0]
    y = points[:, 1]

    g = np.sin(4*np.pi * x) + np.sin(4*np.pi * y)
    return g


def harmonic_rhs(point):
    x = point[0]
    y = point[1]

    f = (x**2 + y**2)*100
    return f

domain_rhs = harmonic_rhs
domain_boundary_condition = harmonic_boundary_condition
```

Se prueban las funciones con las distintas Polyllas

```python
folder = "polyllaMeshes/"
#NumPoints = 10
#NumPoints = 500
NumPoints = 2500
```

```python
#uniform
u = vem(folder + str(NumPoints) + "uniform.polylla.mat", domain_rhs, domain_boundary_condition)
plot_solution(folder + str(NumPoints) + "uniform.polylla.mat", u)
```

```python
#semiUniform
u = vem(folder + str(NumPoints) + "semiUniform.polylla.mat", domain_rhs, domain_boundary_condition)
plot_solution(folder + str(NumPoints) + "semiUniform.polylla.mat", u)
```

```python
# Poisson
u = vem(folder + str(NumPoints) + "poisson.polylla.mat", domain_rhs, domain_boundary_condition)
plot_solution(folder + str(NumPoints) + "poisson.polylla.mat", u)
```

```python
#random
u = vem( folder + str(NumPoints) +  "random.polylla.mat", domain_rhs, domain_boundary_condition)
plot_solution(folder + str(NumPoints) + "poisson.polylla.mat", u)
```
